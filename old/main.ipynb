```python
import pandas as pd
import scipy as spy
import ast
from pyclustering.cluster.cure import cure
from pyclustering.utils import read_sample
from pyclustering.cluster import cluster_visualizer_multidim

print("hey")
```

```python

data = pd.read_excel("data.xlsx")
#rename index
data.rename(columns = {"Unnamed: 0":"main_index"}, inplace=True)
names=["Phi0", "Phi1"]
#transform Phi column into a column of tuples
data["Phi"]=data["Phi"].apply(ast.literal_eval)

#separate Phi into two column
for index,column in enumerate(names):
    data[column] = data['Phi'].apply(lambda Phi: Phi[index])

data.loc[:2,["Phi","Phi0","Phi1"]]
```

```python
#show some data of the dataframe

print ("indexes :\n", list(data.keys()),"\n")
print("\nUnique Values\n")
for column in data.keys()[5:-10]:
    print(f"{column}:\n{data[column].unique()}\n")
```

```python
#grouping the dataframe
group_names = ['Experiment Type', 'Reactor', 'Target', 'Fuels', 'Phi', 'Pressure (Bar)', 'Temperature (K)', 'd0L2', 'd1L2', 'd0Pe', 'd1Pe', 'shift', 'Phi0', 'Phi1']
grouped = data.groupby(group_names)
groups = [x.reset_index() for _, x in grouped]
for i in groups:
    print(i.loc[:,["main_index", "Experiment Type", "Target", "Fuels"]])
    print(type(i),"\n\n")
```

```python
print(groups[0].loc[:,["Phi", "d0L2", "d1L2", "d0Pe", "d1Pe"]].to_numpy())
```

```python
#Example of pyclustering library

input_data = list(groups[-1].loc[:,["Phi0","Phi1","d0L2", "d1L2", "d0Pe", "d1Pe"]].to_numpy())
# Allocate clusters.
cure_instance = cure(input_data, 4);
cure_instance.process();
clusters = cure_instance.get_clusters();
 
# Visualize allocated clusters.
visualizer = cluster_visualizer_multidim()
visualizer.append_clusters(clusters, input_data)
visualizer.show()
```

```python
def cure_algo(data, n_clusters, columns, show=False):
    
    input_data = list(data.loc[:,columns].to_numpy())
    # Allocate three clusters.
    cure_instance = cure(input_data, n_clusters);
    cure_instance.process();
    clusters = cure_instance.get_clusters();
 
    # Visualize allocated clusters.
    if show==True:
        visualizer = cluster_visualizer_multidim()
        visualizer.append_clusters(clusters, input_data)
        visualizer.show()
    return clusters
```

```python
print(f"total number of groups:{len(groups)}")
data_names = ["Phi0","Phi1","d0L2", "d1L2", "d0Pe", "d1Pe"]

#change show to True to display the clusters, warning: could be a lot of computing :)
clustered_groups = [[
    cure_algo(group, n_clusters, data_names, show = False)  for n_clusters in range(1,10)] 
                                                            for group in groups]
print(len(clustered_groups))
```
